```
def getLength(db_name):
  dbname = '{}_crop'.format(db_name)
  ages = os.scandir(dbname)
  length = 0
  for age in ages:
    if '.mat' not in str(age.path):
      length+=len(os.listdir(age.path))
  return length
#print(getLength('imdb'))
print(getLength('wiki'))
#print('total images/samples: ' ,getLength('imdb') + getLength('wiki'))
```

```
# fetch .mat file and load data
def getData(db_name):
  directory = '{}_crop'.format(db_name)
  filename =  '{}.mat'.format(db_name)
  comp_name = os.path.join(directory,filename)
  dbMat = scipy.io.loadmat(comp_name)
  dbPlace = dbMat[db_name][0][0]
  place = dbPlace
  where = comp_name
  return place
```

```
def cleaningData(db_name,place):
  newDirectory = 'new{}'.format(db_name) # making directory name similar to the db used
  try:
    os.makedirs(newDirectory)
  except OSError:
    pass
  pic_num = 0 # to rename the images number wise 
  imList = list() # total Images
  corImList = 0 # correctly classified and fine Images
  failedList = 0 # mislabelled or destroyed Images

  dbname = '{}_crop'.format(db_name)

  for folders in glob(dbname + '/*'):
    for img in glob('%s/*'%folders):
      a = img # asssigning image name to a 
      imList.append(img)  # adding names in a list for personal satisfaction
      pname,fname,pth = a.split('/') # splitting picture's name at /
      bYear = int(place[0][0][pic_num]/365) #birth year
      taken = place[1][0][pic_num] #photo taken
      gender = place[3][0][pic_num] # Female/Male
      faceScore = place[6][0][pic_num] #Face score
      secFaceScore = place[7][0][pic_num] #Sec face score
      age = taken - bYear #calculating age
      if 'n' not in str(faceScore):
        if 'a' not in str(secFaceScore):
          if math.isnan(gender):
            print("Filtered Out")
          else:
            gender = int(gender)
            files = '{}_{}_{}_{}'.format(age,gender,pname,pic_num)
            shutil.copy(img, os.path.join(newDirectory, files))
            corImList += 1
      pic_num += 1
```

```
def startCleaning(db_name):
  getLength(db_name)
  place = getData(db_name)
  cleaningData(db_name,place)
startCleaning('wiki')
```
